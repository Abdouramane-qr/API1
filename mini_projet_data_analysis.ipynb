##3.Appliquer les opérations de prétraitement et filtres nécessaires à ce fichier, pour ne garder que les colonnes correspondantes



* Identifiants et Renommage



```python
# Colonnes officielles du format Geonames
columns = [
    "geonameid", "name", "asciiname", "alternatenames", "latitude", "longitude",
    "feature_class", "feature_code", "country_code", "cc2", "admin1_code",
    "admin2_code", "admin3_code", "admin4_code", "population", "elevation",
    "dem", "timezone", "modification_date"
]

# Assigner les noms de colonnes
df.columns = columns

#Filter des colonnes necessaires
df_filtered = df[["geonameid", "name", "latitude", "longitude"]]

#Renommer les colonnes
df_filtered.columns = ["ID", "location_name", "lat", "long"]
df_filtered.head()

```



*Sauvegarder ces données dans un fichier CSV, nommez-le burkina_location.csv




```python
df_filtered.to_csv(r'/content/drive/MyDrive/Mini_Projet/burkina_location.csv',
    index=False,
    encoding="utf-8")
df_filtered.head()
```

##4 - Opérations sur le fichier CSV burkina_location.csv.



* extraction des données contenant le nom 'gounghin', enregistrez-le sous le fichier gounghin.csv





```python
import pandas as pd

#Chargeement du fichier
df_filtered = pd.read_csv(r'/content/drive/MyDrive/Mini_Projet/burkina_location.csv')

df_gounghin = df_filtered[df_filtered['location_name'].str.contains('gounghin', case=False, na=False)]

df_gounghin.to_csv(r'/content/drive/MyDrive/Mini_Projet/gounghin.csv', index=False)

df_gounghin.head()


```



* extraiction de la sous-partie de la base de données (fichier burkina_location.csv), dont les noms les premières lettres des noms de lieux sont compris entre 'A' et 'P' (ordre alphabétique)


```python
df_A_P = df_filtered[df_filtered['location_name'].str[0].str.upper().between('A', 'P')]

df_A_P.to_csv(r'/content/drive/MyDrive/Mini_Projet/A_P.csv', index=False)

df_A_P.head()
```



* Identification respectivement, la latitude, la longitude minimale et les noms de lieux correspondants


```python
min_lat = df_filtered['lat'].min()
min_lat_name = df_filtered[df_filtered['lat']==min_lat]['location_name'].tolist()

min_long = df_filtered['long'].min()
min_long_name = df_filtered[df_filtered['long']==min_long]['location_name'].tolist()

print(f"Latitude Minimale: {min_lat}")
print(f"Nom des Lieux correspondants: {', '.join(min_lat_name)}")

print(f"Longitude Minimale: {min_long}")
print(f"Nom des Lieux correspondants: {', '.join(min_long_name)}")
```



*  les lieux dont les coordonnées sont comprises entre (lat >= 11 et lon <= 0.5)




```python
df_coord= df_filtered[(df_filtered['lat']>=11) & (df_filtered['long']<=0.5)]
df_coord.to_csv(r'/content/drive/MyDrive/Mini_Projet/coord.csv', index=False)


print("Lieux avec lat >= 11 et lon <= 0.5 :\n", df_coord["location_name"].tolist())

df_coord.head()
```
